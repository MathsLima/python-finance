```python
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go
```

```python
ticker = yf.Ticker('ITUB4.SA')

#intervalo
df = ticker.history(period='1y')
df
```

```python
#calculo de media movel
df['MM5dias'] = df['Close'].rolling(window=5).mean().round(2) #rolling calcula a media movel e window é o delta de tempo
df['MM15dias'] = df['Close'].rolling(window=15).mean().round(2)
```

```python
#grafico com os dados das ações
candlestick = go.Candlestick(
        x=df.index,
        open=df['Open'],
        high=df['High'],
        low=df['Low'],
        close=df['Close']
        )

#grafico disperção da media de 5 dias
lines_mm5 = go.Scatter(x=df.index,
                       y=df['MM5dias'],
                       mode = 'lines',
                       line_color = 'blue',
                       name='MM5d')

#grafico disperção da media de 15 dias
lines_mm15 = go.Scatter(x=df.index,
                       y=df['MM5dias'],
                       mode = 'lines',
                       line_color = 'yellow',
                       name='MM15d')

#figura unindo os graficos
fig = go.Figure(data=[candlestick])
fig.add_trace(lines_mm5)
fig.add_trace(lines_mm15)

fig.show()
```

```python
#media movel exponencial
df['MME25dias'] = df['Close'].ewm(span=25).mean()
```

```python
#cruzando duas medias moveis exponenciais
df['MME30dias'] = df['Close'].ewm(span=30).mean()
df['MME100dias'] = df['Close'].ewm(span=100).mean()
```

```python
#grafico media movel exponencial
candlestick = go.Candlestick(
                           x=df.index,
                           open=df['Open'],
                           high=df['High'],
                           low=df['Low'],
                           close=df['Close']
)

#grafico disperção da media de 5 dias
lines_mme30 = go.Scatter(x=df.index,
                       y=df['MME30dias'],
                       mode = 'lines',
                       line_color = 'blue',
                       name='MME30d')

#grafico disperção da media de 15 dias
lines_mme100 = go.Scatter(x=df.index,
                       y=df['MME100dias'],
                       mode = 'lines',
                       line_color = 'yellow',
                       name='MM100d')

#figura unindo os graficos
fig = go.Figure(data=[candlestick])
fig.add_trace(lines_mme30)
fig.add_trace(lines_mme100)

fig.show()
```
