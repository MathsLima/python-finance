```python
import pandas as pd
import numpy as np
import matplotlib as plt
import pandas_datareader.data as web
import yfinance as yf
import seaborn as sns
```

```python
tickers = ["^BVSP", "USDBRL=X"]
carteira =  yf.download(tickers, start="2018-01-01",)['Close']
```

```python
carteira = carteira.dropna()
carteira.columns = ['Dolar', 'Ibovespa']
```

```python
carteira.plot(subplots=True, figsize=(22,8))
```

```python
sns.heatmap(carteira.corr(), annot=True)
```

```python
carteira['Dolar'].rolling(252).corr(carteira['Ibovespa']).plot(figsize=(22,8))
```

```python
carteira['Ibovespa Dolarizado'] = (carteira['Ibovespa'] / carteira['Dolar'])
```
